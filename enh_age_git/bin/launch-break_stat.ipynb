```python
def sbatch_age_sequence(file, run_any):
    cmd = "sbatch /dors/capra_lab/users/fongsl/enh_age/bin/break_stats.slurm %s"% (file)
        
    print("SLURM", file, datetime.datetime.now())
    print(cmd)
    if run_any ==1:
        subprocess.call(cmd, shell=True)
        
def python_age_sequence(file, run_any):
    cmd = "python /dors/capra_lab/users/fongsl/enh_age/bin/break_stats.py %s"% (file)

    print("PYTHON", file, datetime.datetime.now())
    print(cmd)
    if run_any ==1:
        subprocess.call(cmd, shell=True)
        
def python_age_sequence_oldest(file, run_any):
    cmd = "python /dors/capra_lab/users/fongsl/enh_age/bin/break_stats-oldest.py %s"% (file)

    print("PYTHON", file, datetime.datetime.now())
    print(cmd)
    if run_any ==1:
        subprocess.call(cmd, shell=True)
```

```python
import glob
import os
import subprocess
PATH = "/dors/capra_lab/projects/enhancer_ages/chen_2018/sorted/shuffle/breaks/"

os.chdir(PATH)
FS = glob.glob( "*.bed")
```

```python
complete = []
```

```python
for F in FS:
    n =(F.split("-sorted-")[1]).split("_enh_age_arch_summary.bed")[0]
    if n not in complete:
        new_n = int(n)+900
        newf = f"shuf-sorted-{new_n}_enh_age_arch_summary.bed"
        os.replace(F, newf)
        complete.append(new_n)
```

```python
AGEPATH = "/dors/capra_lab/projects/enhancer_ages/chen_2018/sorted/shuffle/ages/"
os.chdir(AGEPATH)
FS = glob.glob( "*.bed")
for F in FS:
    n =(F.split("-sorted-")[1]).split("_ages.bed")[0]
    if n not in complete:
        new_n = int(n)+900
        newf = f"shuf-sorted-{new_n}_ages.bed"
        os.replace(F, newf)
        complete.append(new_n)

```

```python
def make_shuf_bed(shuf, new_shuf):
    
    cmd = f"cut -f 1,2,3 {shuf} > ../{new_shuf}"
    subprocess.call(cmd, shell = True)
    print(cmd)
    
```

```python
for F in FS: 
    new_shuf = F.split("_enh_age_arch_summary.bed")[0] +".bed"

    make_shuf_bed(F, new_shuf)
```

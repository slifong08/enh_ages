# functions

```python
def open_format_info(bed):
    info = pd.read_csv(bed, sep= '\t', header = None)

    cols = ["#chr", "start", "end", "region_id", 
            "#chr_enh", "start_enh", "end_enh", 
            "seg_index", 'core_remodeling', 'core', "mrca"
           ]
    info.columns = cols  # rename columns 
    info['start'] +=1 #change the start position because msa_split does stuff. 
    info["region_id"] = info["#chr"]+":"+info["start"].map(str)+"-"+ +info["end"].map(str) # redo region id w new start
    return info[["#chr", "start", "end", "region_id", 'core_remodeling', 'core', "mrca"]].drop_duplicates()
```

```python
def custom_round(x, base=10):
    return int(base * round(float(x)/base))

def match_len(Df1, Df2, base_len):
    """
    match df2 on df1 sequence lengths, rounding to the nearest length (e.g. rounding to 10 bp)
    return list of region_ids in df2 matched to length distribution in df1. 
    
    args 
    
    df1: pandas dataframe w/ columns "region id" and "len". Match df2 to df1 length distributions
    df2: pandas dataframe w/ columns "region id" and "len". Match df2 to df1 length distributions
    base_len: int, sequence length range that should be matched on in distribution.
    
    
    process
    
    1. keep only region_id, len in df1, df2
    2. round region lengths to nearest base_len
    3. find overlapping length range for df1, df2
    4. per length, 
    5. get a list of N sample ids in df1, df2 that match length. 
    6. determine min number of region ids w/ match length to select from df1, df2. 
    7. If there are more than N sample ids w matched length in df1 than df2 (or vice versa), 
        randomly select w/o replacement region ids from that dataframe.  
    8. append df1, df2 matching length ids.
    9 return lists of id for df1, df2
    
    notes
    
    - base_len gives better power to match sequences becuase it doesn't require that two sequence match on exact bp lengths, 
        but within a base_len range. E.g. if base_len is 10, start and stop coordinates will be rounded to nearest base10 value.
    

    """
    columns = ["region_id", "len"]
    columns_names = ["matching_ids", "matching_len"]

    #1 df1
    df1 = Df1[columns].drop_duplicates()  # reduce Df1 to only ids and lengths
    df1.columns = columns_names  # rename columns
    
    # same for df2
    df2 = Df2[columns].drop_duplicates()
    df2.columns = columns_names
    
    #2 round df1, df2 to the nearest base_len bps
    df1["matching_len"] = df1["matching_len"].astype(float).apply(lambda x: custom_round(x, base=base_len)) 
    df2["matching_len"] = df2["matching_len"].astype(float).apply(lambda x: custom_round(x, base=base_len))
    
    #3 find intersecting lengths
    lens = set(list(set(df1["matching_len"])) + list(set(df2["matching_len"])))

    df1_match_list, df2_match_list = [], []
    
    #4 for each length, 
    for i, length in enumerate(lens):
        
        #5 count how many regions have length of one size (e.g. 100bp) in df1, df2
        Ndf1 = df1.loc[df1["matching_len"] == length].size
        Ndf2 = df2.loc[df2["matching_len"] == length].size

        #6 get the minimum number of regions to sample from df1, df2. 
        # naturally, n regions of a length for one df will be chosen, while n regions for the other df will be randomly sampled
        # Ns = number to sample
        Ns = min(Ndf1, Ndf2)

        #7 get region ids w/ matching lengths. 
        # Sample randomly from df1, df2 
        # without replacement
        
        if length > 0 and Ns > 0:
            
            df1_ids = df1.loc[df1["matching_len"] == length, "matching_ids"].sample(n = Ns, replace = True) # sample w/ replacement
            df2_ids = df2.loc[df2["matching_len"] == length, "matching_ids"].sample(n = Ns, replace = True) # sample w/ replacement
            
            #8 append list of ids
            df1_match_list.extend(list(set(df1_ids))), df2_match_list.extend(list(set(df2_ids)))

    #9 return matched ids
    return df1_match_list, df2_match_list
```

# load shuffles

```python
# shuffles

samples = [12, 15, 28, 43, 58, 59, 62, 74, 9, 80, 96]

shuf_dict = {}

for s in samples:
    for n in np.arange(1,22):
    
        CHR = f"chr{n}"

        shuf = os.path.join(DATA_PATH, f"tissues-{s}", f"{CHR}_seqid.tsv") # get path
        if os.path.exists(shuf) is True:
            shufDf = pd.read_csv(shuf, sep = '\t') # read file
            shufDf["id"] = f'shuf-{s}_{CHR}' # annotate
            shufDf["erna"] = 0

            shuf_info= os.path.join(SHUF_PATH, f"no-exon_shuf-all_fantom_enh_112_tissues-{s}_age_breaks.bed")

            shufinfo = open_format_info(shuf_info)
            shufinfo = pd.merge(shufDf, shufinfo, how = "left")
            shufinfo["len"] = shufinfo["end"]-shufinfo["start"]

            shuf_dict[f'shuf-{s}_{CHR}']= shufinfo # add df to dictionary
        else:
            print("rerun", shuf)

shuf = pd.concat(shuf_dict.values())
shuf.head()
```

# core, derived v. shuffle

```python
test.groupby(["core","erna"])["percent_id"].describe()["std"].reset_index()['std']
```

```python
col = "mrca_2"
ages = list(cores.sort_values(by = "mrca", ascending =False)[col].round(3).unique())
results = {}
for i, m in enumerate(ages):
    
    test = cores.loc[cores[col].round(3)==round(m, 3)].drop_duplicates()
    
    shuf_cores = test.loc[(test["erna"] == False)& (test["core"]==1), "percent_id"]
    shuf_der = test.loc[(test["erna"] == False) & (test["core"]==0), "percent_id"]
    
    core = test.loc[(test["erna"] == True)& (test["core"]==1), "percent_id"]
    
    der = test.loc[(test["erna"] == True) & (test["core"]==0), "percent_id"]
    
    cs, cp = stats.ttest_ind(core, shuf_cores, equal_var = False)
    ds, dp = stats.ttest_ind(der, shuf_der, equal_var = False)
    #print(test.groupby(["core","erna"])["percent_id"].describe()["std"].reset_index())
    #print(m, test.groupby(["core","erna"])["percent_id"].describe()["S"])
    #print( "\n\n core", cs, cp, "\n\n der",  ds, dp, "\n\n")
    
    newdf = pd.DataFrame({
        "sequence age":[m,m],
        "category":["core", "derived"],
        
        "median sequence identity - eRNA":[np.median(core), np.median(der)],
        "std sequence identity - eRNA":[np.std(core), np.std(der)],
        "median sequence identity - shuffle":[np.median(shuf_cores), np.median(shuf_der)],
        "std sequence identity - shuffle":[np.std(shuf_cores), np.std(shuf_der)],
        "Welch's":[cs, ds],
        "p-value":[cp, dp],
        "N eRNA":[len(core), len(der)],
        "N shuffle":[len(shuf_cores), len(shuf_der)]
    })
    results[i] = newdf

# concat results and save table. 
table = pd.concat(results.values())
table=table.sort_values(by = ["category", "sequence age"], ascending = [True, False])
table.to_csv(os.path.join(RE, "med_age_seqid_erna_v_shuf.tsv"), sep = '\t', index = False)
```

are the derived sequences of this age more similar in sequence identity to the cores of the previous age?

last age derived are statistically significant from current age derived. 

```python
col = "mrca_2"
ages = list(cores.sort_values(by = "mrca", ascending =False)[col].round(3).unique())
for i, age in enumerate(ages):
    
    last_age_val = ages[i-1]
    if last_age_val != 0:
    
        current_age = cores.loc[cores[col].round(3)==round(age, 3)]
        last_age = cores.loc[cores[col].round(3)==round(last_age_val, 3)]

        last_cores = list(last_age.loc[(last_age["erna"] == False)& (last_age["core"]==1), "percent_id"])
        last_der = list(last_age.loc[(last_age["erna"] == False) & (last_age["core"]==0), "percent_id"])

        core = list(current_age.loc[(current_age["erna"] == True)& (current_age["core"]==1), "percent_id"])

        der = list(current_age.loc[(current_age["erna"] == True) & (current_age["core"]==0), "percent_id"])

        """
        current age core v. last age core
        current age derived v. last age derived
        
        current age core v. last age derived
        current age derived v. last age core
        """
        cs, cp = stats.ttest_ind(core, last_cores, equal_var = False) 
        ds, dp = stats.ttest_ind(der, last_der, equal_var = False)
        
        cvds, cvdp = stats.ttest_ind(core, last_der, equal_var = False)
        dvcs, dvcp = stats.ttest_ind(der, last_cores, equal_var = False)
        
        current_erna = current_age.loc[current_age["erna"] ==1]
        last_erna = last_age.loc[last_age["erna"] ==1]
        print("last age", last_age_val, "this_age", age,"\n")
        print("this_age",age,current_erna.groupby(["core"])["percent_id"].describe().reset_index()[["core", "50%"]])
        print("last age",last_age_val,last_erna.groupby(["core"])["percent_id"].describe().reset_index()[["core", "50%"]])
        print( "\n\n this v. last core", cs, cp, "\n\n this v. last der",  ds, dp, "\n\n"\
             "\n\n this core v. last der",  cvds, cvdp, "\n\n this der v. last core", dvcs, dvcp, "\n\n")
```
